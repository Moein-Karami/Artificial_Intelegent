            <br>

            <br>
            ۱) مامور می‌تواند در هر مرحله به خانه ی بالا، چپ، راست، پایین (در صورتی که شرایط مسئله را نقض نکند) حرکت کند.

            ۲) اگر مامور در خانه ای باشد که یک محموله نیز در آن خانه است می‌تواند محموله را نیز با خود به یکی از خانه های همسایه منتقل کند.
            <br>
            <br>
            <b>Transition Model:</b>
            <br>

            <br>

            <br>
            <br>
            <b>Goal State:</b>
            در استیت هدف باید تمام محموله ها و مامور در مقصد خود قرار داشته باشند.

```python
trackers = []
packets = []
directions = [[0, 1], [0, -1], [1, 0], [-1, 0]] # all possible move dircections
directions_symbol = ['R', 'L', 'D', 'U']

class Coordinate:
    def __init__(self, listed):
        self.x, self.y = map(int, listed)
        
    def __hash__(self):
        return hash((self.x, self.y))
    
    def __eq__(self, other):
        return (self.x, self.y) == (other.x, other.y)


class Tracker:
    def __init__(self, listed):
        self.coordinate = Coordinate(listed[: 2])
        self.power = int(listed[2])
        self.number = int(listed[3])
    
    def __hash__(self):
        return hash((self.coordinate, self.power, self.number))
    
    def __eq__(self, other):
        return (self.coordinate, self.power, self.number) == (other.coordinate, other.power, other.number)

    
class MovingObject:
    def __init__(self, coordinate, destination, number):
        self.coordinate = coordinate
        self.destination = destination;
        self.number = number

    def __hash__(self):
        return hash((self.coordinate, self.destination, self.number))
    
    def __eq__(self, other):
        return (self.coordinate, self.destination, self.number) == (ohter.coordinate, other.destination, other.number)
    
    
class State:
    def __init__(self, agent, packets, steps_in_tracker_zone = {}, steps = ''):
        self.agent = agent
        self.packets = packets
        self.steps_in_tracker_zone = steps_in_tracker_zone
        self.steps = steps
    
    def __eq__(self, other):
        return (self.agent, self.packets, self.steps_in_tracker_zone, self.steps) == (other.agent, other.packets, other.steps_in_tracker_zone, other.steps)
    
    def __hash__(self):
        return hash((self.agent, map(hash, self.packets), map(hash, self.steps_in_tracker_zone)))
    
    def is_valid(self): #is this state valid or not?
        if self.agent.coordinate.x >= rows or self.agent.coordinate.y >= cols or self.agent.coordinate.x < 0 or self.agent.coordinate.y < 0:
            return False
        for key, value in self.steps_in_tracker_zone:
            if value > trackers[key].power:
                return False
        for i in range(len(self.packets)):
            for j in range(i + 1, len(self.packets)):
                if self.packets[i].coordinate == self.packets[j].coordinate:
                    return False
    
    def get_new_states(self): #returns states which can be reached from current state
        new_states = []
        for direct, symbol in directions, directions_symbol:
            new_state = self
            new_state.agent.coordinate = Coordinates([self.agent.coordinate.x + direct[0], self.agent.coordinate.y + direct[1]])
            new_state.steps += symbol
            
            for tracker, steps in self.steps_in_tracker_zone:
                if tracker in field[new_state.agent.coordinate.x, new_state.agent.coordinate.y]:
                    self.steps_in_tracker_zone[tracker] += 1
                else:
                    self.steps_in_tracker_zone.pop(tracker, None)
            
            for tracker in field[new_state.agent.coordinate.x, new_state.agent.coordinate.y]:
                if tracker not in self.steps_in_tracker_zone.keys():
                    self.steps_in_tracker_zone[tracker] = 1
            
            if new_state.is_valid() is False :
                continue
            new_states.append(new_state)
            
            for packet in self.packets:
                if packet.coordinate == self.agent.coordinate:
                    packet.coordinate = new_state.agent.coordinate
            
            if new_state.is_valid():
                new_states.append(new_state)
        return new_state
```
