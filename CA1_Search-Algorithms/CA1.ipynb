```python
trackers = []
packets = []
directions = [[0, 1], [1, 0], [-1, 0], [0, -1]] # all possible move dircections
directions_symbol = ['R', 'D', 'U', 'L']

class Coordinate:
    def __init__(self, listed):
        self.x, self.y = map(int, listed)
        
    def __hash__(self):
        return hash((self.x, self.y))
    
    def __eq__(self, other):
        return (self.x, self.y) == (other.x, other.y)
    
    def __str__(self):
        return f"({self.x}, {self.y})"
    
    def __add__(self, other):
        return Coordinate([self.x + other.x, self.y + other.y])

class Tracker:
    def __init__(self, listed):
        self.coordinate = Coordinate(listed[: 2])
        self.power = int(listed[2])
        self.number = int(listed[3])
    
    def __hash__(self):
        return hash((self.coordinate, self.power, self.number))
    
    def __eq__(self, other):
        return (self.coordinate, self.power, self.number) == (other.coordinate, other.power, other.number)

    def __str__(self):
        return f"Coordinate: {self.coordinate}\npower: {self.power}\nId: {self.number}"

    
class Packet:
    def __init__(self, coordinate, destination, number):
        self.coordinate = coordinate
        self.destination = destination;
        self.number = number

    def __hash__(self):
        return hash((self.coordinate, self.destination, self.number))
    
    def __eq__(self, other):
        return (self.coordinate, self.destination, self.number) == (other.coordinate, other.destination,\
                                                                    other.number)
    
    def __str__(self):
        return f"Id: {self.number}\nCoordinate: {self.coordinate}\ndestination: {self.destination}"

    
class State:
    def __init__(self, agent = Coordinate([0, 0]), remaining_packets = None, steps_in_tracker_zone = None, steps = 0,\
                 carrying_packet_number = None, parent = None, direction = None, number = 0):
        self.agent = agent
        self.remaining_packets = remaining_packets
        self.steps_in_tracker_zone = steps_in_tracker_zone
        self.steps = steps
        self.carrying_packet_number = carrying_packet_number
        self.parent = parent
        self.direction = direction
        self.number = number
    
    def to_string(self):
        string = f"agent: {self.agent}\npackets: "
        string += ' '.join(map(str, self.remaining_packets))
        string += f"\ncarrying packet number: {self.carrying_packet_number}\n"
        string += f"trackers: {self.steps_in_tracker_zone}\n"
        return string
    
    def __eq__(self, other):
        return (self.agent, self.remaining_packets, self.carrying_packet_number,\
                self.steps_in_tracker_zone) == (other.agent, other.remaining_packets,\
                                                other.carrying_packet_number, other.steps_in_tracker_zone)
    
    
    def __hash__(self):
        return hash((self.agent, str(self.carrying_packet_number), sum(self.remaining_packets),\
                     str(self.steps_in_tracker_zone)))
    
    def __str__(self):
        string = "---------\n" + self.to_string()
        string += f"steps:  {str(self.steps)}"
        return string + '\n---------\n'
    
    def __lt__(self, other): #returns true if self.steps is less than other.steps
        return self.steps < other.steps
    
    def fix_carrying_packet(self): #put packet down if we are in its destination and pull it up if 
        #we are in the same place with a packet
        if self.carrying_packet_number != None:
            if self.agent == packets[self.carrying_packet_number].destination:
                self.remaining_packets.remove(self.carrying_packet_number)
                self.carrying_packet_number = None

        if self.carrying_packet_number == None:
            for packet in self.remaining_packets:
                if packets[packet].coordinate == self.agent:
                    self.carrying_packet_number = packet
                    break
    
    def update_steps_in_tracker_zone(self): #returns false if some tracker tracks us (we have stayed in
        #some tracker zone too much)

        if self.steps_in_tracker_zone != None:
            if self.steps_in_tracker_zone[0] == field[self.agent.x, self.agent.y]:
                self.steps_in_tracker_zone[1] += 1
                if self.steps_in_tracker_zone[1] > trackers[self.steps_in_tracker_zone[0]].power:
                    return False
            else:
                self.steps_in_tracker_zone = None

        if self.steps_in_tracker_zone == None and field[self.agent.x, self.agent.y] != None:
            self.steps_in_tracker_zone = [field[self.agent.x, self.agent.y], 1]
        return True

    
    def get_new_states(self): #returns states which can be reached from current state
        new_states = []
        
        for direct, symbol in zip(directions, directions_symbol):
            if self.agent.x + direct[0] >= rows or self.agent.y +\
            direct[1] >= cols or self.agent.x + direct[0] < 0 or self.agent.y\
            + direct[1] < 0:
                continue
   
            new_state = State(remaining_packets =\
                              copy.deepcopy(self.remaining_packets), steps = self.steps,\
                              steps_in_tracker_zone = copy.deepcopy(self.steps_in_tracker_zone),\
                              carrying_packet_number = self.carrying_packet_number)

            new_state.agent = self.agent + Coordinate(direct)
            new_state.steps += 1
            new_state.direction = symbol
            new_state.parent = self.number

            if new_state.update_steps_in_tracker_zone() is False:
                continue
            new_state.fix_carrying_packet()
            new_states.append(new_state) 
        return new_states
    
    def is_goal(self): #returns true if current state is goal state
        if self.agent != agent_destination:
            return False
        if self.remaining_packets != []:
            return False
        return True
    
    def get_T(self): #return T(x)
        return (len(self.remaining_packets) - 1) * min_packet_dest_to_start
    
    def get_h(self): #returns value of heuristic function value
        if len(self.remaining_packets) == 0:
            return get_distance(self.agent, agent_destination)
        min_dist = float('inf')
        sum_dist = 0
        for packet in self.remaining_packets:
            if packet != self.carrying_packet_number:
                sum_dist += get_distance(packets[packet].coordinate, packets[packet].destination)
                min_dist = min(min_dist, get_distance(self.agent, packets[packet].coordinate))
            else:
                sum_dist += get_distance(self.agent, packets[packet].destination)
                min_dist = 0
        return min_dist + sum_dist + self.get_T() + min_dist_to_dest
```

```python
def get_distance(loc1, loc2):
    return abs(loc1.x - loc2.x) + abs(loc1.y - loc2.y)

def get_path(cur_state, explored): #returns moves which is done to reach to cur_state
    path = ''
    while cur_state.parent is not None:
        path += cur_state.direction
        cur_state = explored[cur_state.parent]
    return path[::-1]
```

```python
file = open("sample_testcases/test_00.txt")
packet_dest_to_start = [] #list of distance of destination and start point of pair of different packets
min_packet_dest_to_start = float('inf') #minimum distance of destination and start point of pair of different packets
min_dist_to_dest = float('inf') #minimum distance between agent destination and packets' destinations

rows, cols = map(int, file.readline().split())

agent = Coordinate(file.readline().split())
agent_destination = Coordinate(file.readline().split())

number_of_trackers, number_of_packets = map(int, file.readline().split())

for i in range(number_of_trackers):
        trackers.append(Tracker(file.readline().split() + [i]))

for i in range(number_of_packets):
        packets.append(Coordinate(file.readline().split()))
for i in range(number_of_packets):
        packets[i] = Packet(packets[i], Coordinate(file.readline().split()), i)

for packet in packets:
    for other_packet in packets:
        if (packet != other_packet):
            packet_dest_to_start.append(get_distance(packet.destination, other_packet.coordinate))
    min_dist_to_dest = min(min_dist_to_dest, get_distance(agent_destination, packet.destination))

packet_dest_to_start.sort()
min_packet_dest_to_start = packet_dest_to_start[0]
    
initial_state = State(agent, [i for i in range(number_of_packets)])
```

```python
field = {}
for i in range(rows):
    for j in range(cols):
        field[i, j] = None
        for tracker in trackers:
            if get_distance(Coordinate([i, j]), tracker.coordinate) <= tracker.power:
                field[i, j] = tracker.number
```

    <div dir = "rtl" style = "font-size: 16px">
            این الگوریتم یک الگوریتم
        uninformed
        است که راس های مختلف را بر اساس ارتفاع آن‌ها می‌بیند و وقتی تمام راس های  موجود در یک ارتفاع را دید، به ارتفاع بعدی رفته و رئوس ارتفاع جدید را می‌بیند. در این الگوریتم همواره استیت ها و راس های مرزی را در یک صف نگه داشته و به ترتیب از سر صف گسترش می‌دهیم. همچنین برای جلوگیری از دیدن راس های تکراری، راس های گسترش داده شده را نیز در یک 
        set
        نگه می‌داریم.
        <br>
        مزایا: این الگوریتم کامل است و جواب بهینه را ارائه می‌دهد  و نسبت به 
        IDS
        سریع تر است.
        <br>
        معایب: از آنجایی که این الگوریتم
        uninformed
        است، به حافظه و زمان زیادی نیاز دارد و از 
        A*
        کند تر است.
        <br>
        Time complexity: O(b^d)
        <br>
        Space complexity: O(b^d)
    </div>
</html>

```python
def bfs(initial_state):
    explored = set()
    frontier = queue.Queue()
    seen_states = 0
    explored_dict = {}
    
    frontier.put(copy.deepcopy(initial_state))

    while frontier.empty() == False:
        cur_state = frontier.get()
        cur_state.number = seen_states
        
        if cur_state in explored:
            continue

        seen_states += 1
        
        if cur_state.is_goal():
            return (cur_state, seen_states, get_path(cur_state, explored_dict))

        explored.add(cur_state)
        explored_dict[cur_state.number] = cur_state

        new_states = cur_state.get_new_states()

        for state in new_states:
            if state not in explored:
                frontier.put(state)

    return (None, seen_states, None)
```

           IDS:
        </b>
    </div>
    <br>
    <div dir = "rtl" style = "font-size: 16px">
            این الگوریتم یک الگوریتم
        uninformed
        است و مانند الگوریتم
        BFS
        کامل و بهینه است ولی نسبت به آن به زمان بیشتری برای محاسبات نیاز دارد اما به حافظه‌ی کمتری نیاز دارد، چون تشابه زیادی به 
       DFS
        دارد و تنها تفاوت آن در این است که هر بار مشخص است تا چه عمقی پیش می‌رویم.
        <br>
        در پیاده سازی زیر تابع 
        ids
        هر بار یک عمق را به عنوان حداکثر عمق پیمایش مشخص می‌کند و این مقدار را به تابع
        dfs
        می‌دهد تا 
        dfs
        روی درخت تا این ارتفاع مشخص شده انجام شود.
        <br>
        مزایا: کامل و اپتیمال است و به فضای کمی نسبت به دو الگوریتم دیگر نیاز دارد.
        <br>
        معایب: نسبت به بقیه الگوریتم ها کند تر است.
        <br>
        Time complexity: O(b^d)
        <br>
        Space complexity: O(b^d)
        <br>
        توجه کنید با اینکه به صورت تئوری پیچیدگی زمانی و حافظه این الگوریتم با 
        BFS
        برابر است ولی در عمل این الگوریتم حافظه ی کمتر و زمان بیشتری می‌خواهد.
    </div>
</html>

```python
def dfs(cur_state, explored, seen_states, explored_dict, max_depth):
    if (cur_state, cur_state.steps) in explored or cur_state.steps > max_depth:
        return None
    cur_state.number = seen_states[0]
    explored_dict[seen_states[0]] = cur_state
    explored.add((cur_state, cur_state.steps))
    seen_states[0] += 1
    if cur_state.is_goal():
        return cur_state
    
    new_states = cur_state.get_new_states()
    
    for state in new_states:
        if (state, state.steps) not in explored:
            goal_state = dfs(state, explored, seen_states, explored_dict, max_depth)
            if goal_state is not None:
                return goal_state
    return None

def ids(initial_state):
    depth = 0
    seen_states = [0]
    while True:
        explored = set()
        explored_dict = {}
        goal_state = dfs(initial_state, explored, seen_states, explored_dict, depth)
        if goal_state is not None:
            break
        depth += 1;
    
    return (goal_state, seen_states, get_path(goal_state, explored_dict))
```

```python
start_time = time.time()
(goal_state, seen_states, path) = ids(initial_state)
finish_time = time.time()

print("Executation Time:", finish_time - start_time, "\nSeen States:", seen_states[0])
# print(goal_state)
print("path: ", path)
print("steps: ", len(path))
```

        <br>
        پس در این حالت هم مقدار تابع حداکثر یک واحد کاهش می‌یابد و حکم اثبات می‌شود.

```python
def A_star(initial_state, weight = 1):
    explored = set()
    frontier = [(initial_state.get_h() * weight, initial_state)]
    seen_states = 0
    explored_dict = {}
    front_set = set()
    
    while len(frontier):
        cur_state = heapq.heappop(frontier)[1]
        cur_state.number = seen_states
        
        if cur_state in explored:
            continue

        seen_states += 1
        
        if cur_state.is_goal():
            return (cur_state, seen_states, get_path(cur_state, explored_dict))

        explored.add(cur_state)
        explored_dict[cur_state.number] = cur_state
        
        new_states = cur_state.get_new_states()
        
        for state in new_states:
            if state not in explored and state not in front_set:
                heapq.heappush(frontier, (state.steps + state.get_h() * weight, state))
                front_set.add(state)

    return (None, seen_states, None)
```
