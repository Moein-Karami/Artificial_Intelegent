            <br>

            <br>
            ۱) مامور می‌تواند در هر مرحله به خانه ی بالا، چپ، راست، پایین (در صورتی که شرایط مسئله را نقض نکند) حرکت کند.
            <br>
            ۲) اگر مامور در خانه ای باشد که یک محموله نیز در آن خانه است می‌تواند محموله را نیز با خود به یکی از خانه های همسایه منتقل کند.
            <br>
            <br>
            <b>Transition Model:</b>
            <br>
            ۱) اگر مامور در هر جهتی حرکت کند، در نهایت به خانه جدید می‌رود.
            <br>
            ۲) اگر مامور بسته را نیز با خود حمل کند، پس از حرکت در هر جهتی، در نهایت هم خودش و هم بسته به خانه جدید منتقل می‌شوند.
            <br>
            <br>
            <b>Goal State:</b>
            در استیت هدف باید تمام محموله ها و مامور در مقصد خود قرار داشته باشند.
            <br>
            <br>
            <b> Path Cost: </b>
            هر عمل و حرکت مامور، ۱ واحد هزینه دارد.

```python
trackers = []
packets = []
directions = [[0, 1], [1, 0], [-1, 0], [0, -1]] # all possible move dircections
directions_symbol = ['R', 'D', 'U', 'L']

class Coordinate:
    def __init__(self, listed):
        self.x, self.y = map(int, listed)
        
    def __hash__(self):
        return hash((self.x, self.y))
    
    def __eq__(self, other):
        return (self.x, self.y) == (other.x, other.y)
    
    def __str__(self):
        return f"({self.x}, {self.y})"


class Tracker:
    def __init__(self, listed):
        self.coordinate = Coordinate(listed[: 2])
        self.power = int(listed[2])
        self.number = int(listed[3])
    
    def __hash__(self):
        return hash((self.coordinate, self.power, self.number))
    
    def __eq__(self, other):
        return (self.coordinate, self.power, self.number) == (other.coordinate, other.power, other.number)

    def __str__(self):
        return f"Coordinate: {self.coordinate}\npower: {self.power}\nId: {self.number}"

    
class MovingObject:
    def __init__(self, coordinate, destination, number):
        self.coordinate = coordinate
        self.destination = destination;
        self.number = number

    def __hash__(self):
        return hash((self.coordinate, self.destination, self.number))
    
    def __eq__(self, other):
        return (self.coordinate, self.destination, self.number) == (other.coordinate, other.destination, other.number)
    
    def __str__(self):
        return f"Id: {self.number}\nCoordinate: {self.coordinate}\ndestination: {self.destination}"

    
class State:
    def __init__(self, agent = None, remaining_packets = None, steps_in_tracker_zone = None, steps = 0, carrying_packet_number = None, parent = None, direction = None, number = 0):
        self.agent = agent
        self.remaining_packets = remaining_packets
        self.steps_in_tracker_zone = steps_in_tracker_zone
        self.steps = steps
        self.carrying_packet_number = carrying_packet_number
        self.parent = parent
        self.direction = direction
        self.number = number
    
    def to_string(self):
        string = f"agent: {self.agent}\npackets: "
        string += ' '.join(map(str, self.remaining_packets))
        string += f"\ncarrying packet number: {self.carrying_packet_number}\n"
        string += f"trackers: {trackers}\n"
        return string
    
    def __eq__(self, other):
        return (self.agent, self.remaining_packets, self.carrying_packet_number, self.steps_in_tracker_zone) == (other.agent, other.remaining_packets, other.carrying_packet_number, other.steps_in_tracker_zone)
    
    
    def __hash__(self):
        return hash((self.agent, str(self.carrying_packet_number), sum(self.remaining_packets), str(self.steps_in_tracker_zone)))
    
    def __str__(self):
        string = "---------\n" + self.to_string()
        string += f"steps:  {str(self.steps)}"
        return string + '\n---------\n'
    
    
    def fix_carrying_packet(self): #put packet down if we are in its destination and pull it up if we are in the same place with a packet
        if self.carrying_packet_number != None:
            if self.agent.coordinate == packets[self.carrying_packet_number].destination:
                self.remaining_packets.remove(self.carrying_packet_number)
                self.carrying_packet_number = None

        if self.carrying_packet_number == None:
            for packet in self.remaining_packets:
                if packets[packet].coordinate == self.agent.coordinate:
                    self.carrying_packet_number = packet
                    break
    
    def update_steps_in_tracker_zone(self): #returns false if some tracker tracks us (we have stayed in some tracker zone too much)
        if self.steps_in_tracker_zone != None:
            if self.steps_in_tracker_zone[0] == field[self.agent.coordinate.x, self.agent.coordinate.y]:
                self.steps_in_tracker_zone[1] += 1
                if self.steps_in_tracker_zone[1] > trackers[self.steps_in_tracker_zone[0]].power:
                    return False
            else:
                self.steps_in_tracker_zone = None

        if self.steps_in_tracker_zone == None and field[self.agent.coordinate.x, self.agent.coordinate.y] != None:
            self.steps_in_tracker_zone = [field[self.agent.coordinate.x, self.agent.coordinate.y], 1]
        return True

    
    def get_new_states(self, last_used_id): #returns states which can be reached from current state
        new_states = []
        
        for direct, symbol in zip(directions, directions_symbol):
            if self.agent.coordinate.x + direct[0] >= rows or self.agent.coordinate.y + direct[1] >= cols or self.agent.coordinate.x + direct[0] < 0 or self.agent.coordinate.y + direct[1] < 0:
                continue
   
            new_state = State(agent = copy.deepcopy(self.agent), remaining_packets = copy.deepcopy(self.remaining_packets), steps = self.steps, steps_in_tracker_zone = copy.deepcopy(self.steps_in_tracker_zone), carrying_packet_number = self.carrying_packet_number)
#             new_state = copy.deepcopy(self)
            new_state.agent.coordinate = Coordinate([self.agent.coordinate.x + direct[0], self.agent.coordinate.y + direct[1]])
            new_state.steps += 1
            new_state.direction = symbol
            new_state.parent = self.number
            last_used_id[0] += 1
            new_state.number = last_used_id[0]

            if new_state.update_steps_in_tracker_zone() is False:
                continue
            new_state.fix_carrying_packet()
            new_states.append(new_state) 
        return new_states
    
    def is_goal(self): #returns true if current state is goal state
        if self.agent.coordinate != self.agent.destination:
            return False
        if self.remaining_packets != []:
            return False
        return True
    
```

```python
def get_distance(loc1, loc2):
    return abs(loc1.x - loc2.x) + abs(loc1.y - loc2.y)

def get_path(cur_state, explored): #returns moves which done to reach to cur_state
    path = ''
    while cur_state.number:
        path += cur_state.direction
        cur_state = explored[cur_state.parent]
    return path[::-1]
```

```python
file = open("sample_testcases/test_00.txt")

rows, cols = map(int, file.readline().split())

agent = MovingObject(Coordinate(file.readline().split()), Coordinate(file.readline().split()), -1)

number_of_trackers, number_of_packets = map(int, file.readline().split())

for i in range(number_of_trackers):
        trackers.append(Tracker(file.readline().split() + [i]))

for i in range(number_of_packets):
        packets.append(Coordinate(file.readline().split()))
for i in range(number_of_packets):
        packets[i] = MovingObject(packets[i], Coordinate(file.readline().split()), i)
        
initial_state = State(agent, [i for i in range(number_of_packets)])
```

    <div dir = "rtl" style = "font-size: 16px">
            برای راحتی کار در ادامه،‌مشخص می‌کنیم که هر خانه از جدول داخل منطقه ی کدام آنتن‌ها است. 
    </div>

```python
field = {}
for i in range(rows):
    for j in range(cols):
        field[i, j] = None
        for tracker in trackers:
            if get_distance(Coordinate([i, j]), tracker.coordinate) <= tracker.power:
                field[i, j] = tracker.number
```

<html>
    <div dir = "rtl" style="font-size: 22px; color: green">
        <b>
           BFS:
        </b>
    </div>
    <br>
    <div dir = "rtl" style = "font-size: 16px">
            این الگوریتم یک الگوریتم
        uninformed
        است که راس های مختلف را بر اساس ارتفاع آن‌ها می‌بیند و وقتی تمام راس های  موجود در یک ارتفاع را دید، به ارتفاع بعدی رفته و رئوس ارتفاع جدید را می‌بیند. در این الگوریتم همواره استیت ها و راس های مرزی را در یک صف نگه داشته و به ترتیب از سر صف گسترش می‌دهیم. همچنین برای جلوگیری از دیدن راس های تکراری، راس های گسترش داده شده را نیز در یک 
        set
        نگه می‌داریم.
        <br>
        مزایا: این الگوریتم کامل است و جواب بهینه را ارائه می‌دهد.
        <br>
        معایب: از آنجایی که این الگوریتم
        uninformed
        است، به حافظه و زمان زیادی نیاز دارد.
        <br>
        Time complexity: O(b^d)
        Space complexity: O(b^d)
    </div>
</html>

```python
def bfs(initial_state):
    explored = set()
    frontier = queue.Queue()
    seen_states = 0
    last_state_number = [0]
    explored_dict = {}
    
    frontier.put(copy.deepcopy(initial_state))
    s = 0
    while frontier.empty() == False:
        cur_state = frontier.get()
        if cur_state in explored:
            continue

        seen_states += 1

        if cur_state.is_goal():
            return (cur_state, seen_states, get_path(cur_state, explored_dict))
               
        explored.add(cur_state)
        explored_dict[cur_state.number] = cur_state

#         start = time.time()
        new_states = cur_state.get_new_states(last_state_number)
#         print (time.time() - start)
        for state in new_states:
            if state not in explored:
                frontier.put(state)

    return (None, seen_states, None)
```

```python
start_time = time.time()
(goal_state, seen_states, path) = bfs(initial_state)
end_time = time.time()

print(end_time - start_time, seen_states)
print(goal_state)
print(path)
```
