            <br>
            ۲) اگر مامور در خانه ای باشد که یک محموله نیز در آن خانه است می‌تواند محموله را نیز با خود به یکی از خانه های همسایه منتقل کند.
            <br>
            <br>
            <b>Transition Model:</b>
            <br>
            ۱) اگر مامور در هر جهتی حرکت کند، در نهایت به خانه جدید می‌رود.
            <br>

            <br>
            <br>
            <b>Goal State:</b>
            در استیت هدف باید تمام محموله ها و مامور در مقصد خود قرار داشته باشند.

```python
trackers = []
packets = []
directions = [[1, 0], [0, -1], [-1, 0], [0, 1]] # all possible move dircections
directions_symbol = ['D', 'L', 'U', 'R']

class Coordinate:
    def __init__(self, listed):
        self.x, self.y = map(int, listed)
        
    def __hash__(self):
        return hash((self.x, self.y))
    
    def __eq__(self, other):
        return (self.x, self.y) == (other.x, other.y)
    
    def __str__(self):
        return f"({self.x}, {self.y})"


class Tracker:
    def __init__(self, listed):
        self.coordinate = Coordinate(listed[: 2])
        self.power = int(listed[2])
        self.number = int(listed[3])
    
    def __hash__(self):
        return hash((self.coordinate, self.power, self.number))
    
    def __eq__(self, other):
        return (self.coordinate, self.power, self.number) == (other.coordinate, other.power, other.number)

    def __str__(self):
        return f"Coordinate: {self.coordinate}\npower: {self.power}\nId: {self.number}"

    
class MovingObject:
    def __init__(self, coordinate, destination, number):
        self.coordinate = coordinate
        self.destination = destination;
        self.number = number

    def __hash__(self):
        return hash((self.coordinate, self.destination, self.number))
    
    def __eq__(self, other):
        return (self.coordinate, self.destination, self.number) == (other.coordinate, other.destination, other.number)
    
    def __str__(self):
        return f"Id: {self.number}\nCoordinate: {self.coordinate}\ndestination: {self.destination}"

    
class State:
    def __init__(self, agent, packets, steps_in_tracker_zone = {}, steps = ''):
        self.agent = agent
        self.packets = packets
        self.steps_in_tracker_zone = steps_in_tracker_zone
        self.steps = steps
    
    def to_string(self):
        string = f"agent: {self.agent}\n\npackets:\n"
        for packet in self.packets:
            string += str(packet) + '\n'
        string += "\ntrackers:\n"
        for key, value in self.steps_in_tracker_zone:
            string += f'{key}: {value}\n'
        return string
    
    def __eq__(self, other):
        return (self.agent, self.packets, self.steps_in_tracker_zone) == (other.agent, other.packets, other.steps_in_tracker_zone)
    
    
    def __hash__(self):
        return hash((self.agent, sum(map(hash, self.packets)), sum(map(hash, self.steps_in_tracker_zone))))
    
    def __str__(self):
        string = "---------\n" + self.to_string()
        string += "\nsteps: " + self.steps + '\n'
        return string + '---------\n'
    
    def is_valid(self): #is this state valid or not?
        if self.agent.coordinate.x >= rows or self.agent.coordinate.y >= cols or self.agent.coordinate.x < 0 or self.agent.coordinate.y < 0:
            return False
        for key, value in self.steps_in_tracker_zone:
            if value > trackers[key].power:
                return False
        return True
    
    def get_new_states(self): #returns states which can be reached from current state
        new_states = []
        transfered_packet_index = None

        for i in range(len(self.packets)):
            packet = self.packets[i]
            if packet.coordinate == self.agent.coordinate and packet.coordinate != packet.destination:
                transfered_packet_index = i
                break
        
        for direct, symbol in zip(directions, directions_symbol):
            new_state = copy.deepcopy(self)
            new_state.agent.coordinate = Coordinate([self.agent.coordinate.x + direct[0], self.agent.coordinate.y + direct[1]])
            new_state.steps += symbol
            
            for tracker, steps in new_state.steps_in_tracker_zone:
                if tracker in field[new_state.agent.coordinate.x, new_state.agent.coordinate.y]:
                    new_state.steps_in_tracker_zone[tracker] += 1
                else:
                    new_state.steps_in_tracker_zone.pop(tracker, None)
            
            if new_state.is_valid() is False :
                continue
            
            for tracker in field[new_state.agent.coordinate.x, new_state.agent.coordinate.y]:
                if tracker not in new_state.steps_in_tracker_zone.keys():
                    new_state.steps_in_tracker_zone[tracker] = 1
            
            new_states.append(copy.deepcopy(new_state))
            
            flag = False
            for packet in new_state.packets:
                if packet.coordinate == new_state.agent.coordinate:
                    flag = True
                    break
            if flag:
                continue
                
            if transfered_packet_index != None:
                print(new_state)
                new_state.packets[transfered_packet_index].coordinate = copy.deepcopy(new_state.agent.coordinate)                
                new_states.append(copy.deepcopy(new_state))
                print(new_state)
                
        return new_states
    
    def is_goal(self): #returns true if current state is goal state
        if self.agent.coordinate != self.agent.destination:
            return False
        for packet in self.packets:
            if packet.coordinate != packet.destination:
                return False
        return True
    
```

           BFS:

    <br>
    <div dir = "rtl" style = "font-size: 16px">
            این الگوریتم یک الگوریتم
        uninformed
        است که راس های مختلف را بر اساس ارتفاع آن‌ها می‌بیند و وقتی تمام راس های  موجود در یک ارتفاع را دید، به ارتفاع بعدی رفته و رئوس ارتفاع جدید را می‌بیند. در این الگوریتم همواره استیت ها و راس های مرزی را در یک صف نگه داشته و به ترتیب از سر صف گسترش می‌دهیم. همچنین برای جلوگیری از دیدن راس های تکراری، راس های گسترش داده شده را نیز در یک 
        set
        نگه می‌داریم.
        <br>
        مزایا: این الگوریتم کامل است و جواب بهینه را ارائه می‌دهد.
        <br>
        معایب: از آنجایی که این الگوریتم
        uninformed
        است، به حافظه و زمان زیادی نیاز دارد.
        <br>
        Time complexity: O(b^d)
        Space complexity: O(b^d)
    </div>
</html>

```python
def bfs(initial_state):
    explored = set()
    frontier = queue.Queue()
    seen_states = 0
    
    frontier.put(copy.deepcopy(initial_state))
    
    while frontier.empty() == False:
        cur_state = frontier.get()
        if cur_state in explored:
            continue

        seen_states += 1
        
#         print (cur_state.agent.coordinate)

        if cur_state.is_goal():
            return (cur_state, seen_states)
        
        explored.add(copy.deepcopy(cur_state))

        new_states = cur_state.get_new_states()
        
        for state in new_states:
            frontier.put(copy.deepcopy(state))

    return (None, seen_states)
```
